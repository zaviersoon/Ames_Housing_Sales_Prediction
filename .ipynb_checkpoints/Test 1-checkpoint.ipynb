### 5. Check Missing Value

```python
# Define Function to create missing value dataframe 

def missing_val(df):
    
    df_miss = df.isna().sum().reset_index()
    df_miss.columns = ['variables', 'miss_val']
    df_miss['percent'] = round((df_miss['miss_val'] / df.shape[0] * 100 ), 2)
    df_miss = df_miss[df_miss['miss_val'] != 0].sort_values('miss_val', ascending = False)
    
    # Total number of variables with missing values
    print(f'\nTotal number of variables with missing values: {df_miss.shape[0]}\n')
    return df_miss
```

#### 5.1 Train Dataset Missing Value

```python
# Display the summary of missing value in train dataset 

missing_val(train)
```

```python
# Display the summary of missing value in test dataset 

missing_val(test)
```

- Interesting, all 113 row of garage_area = 0.0 prove that all garage_yr_blt, garage_cond, garage_qual, garage_finish, garage_type variable is missing value.
- This confirm that 113 row of garage_yr_blt, garage_finish, garage_qual, garage_cond & garage_type variable missing value should refer to houses without garage.

```python
# Replace 113 row garage_finish, garage_qual, garage_cond & garage_type variable missing value with 'No Garage' using index reference from garage_area_0 = 0.0

train.loc[garage_area_0.index, ['garage_finish', 'garage_qual', 'garage_cond', 'garage_type']] = 'No Garage'

# Replace 113 row garage_yr_blt variable missing value with 0 integer value to represent 'No Garage' using index reference from garage_area = 0.0

train.loc[garage_area_0.index, 'garage_yr_blt'] = 0
```

```python
# Display the updated list of garage related variable number of missing value

garage = train.filter(regex = 'garage')

garage.isnull().sum().sort_values(ascending = False)
```

```python
# Display the row which still has the remaining missing value for garage related variable
# Using .any() to return whether any element is True, at vertical axis.

garage[garage.isnull().any(axis = 1)]
```

- Index 1712 house come with garage however all other garage related variable is missing value!!! The missing value can be replaced with mean/mode value since is just 0.05% of the data.

```python
# Mean of garage_yr_blt
print(f"The mean of garage_yr_blt variable is: {int(round(train['garage_yr_blt'].mean(),0))}")

# Mode of garage_finish
print(f"The mode of garage_finish variable is: {train['garage_finish'].mode()[0]}")

# Mode of garage_cars
print(f"The mode of garage_cars variable is: {train['garage_cars'].mode()[0]}")

# Mean of garage_area
print(f"The mean of garage_area variable is: {int(round(train['garage_area'].mean(), 0))}")

# Mode of garage_qual
print(f"The mode of garage_qual variable is: {train['garage_qual'].mode()[0]}")

# Mode of garage_cond
print(f"The mode of garage_cond variable is: {train['garage_cond'].mode()[0]}")
```

```python
# Replace Index 1712 house missing value with mean/mode

train.loc[1712, 'garage_yr_blt'] = 1870
train.loc[1712, 'garage_finish'] = 'Unf'
train.loc[1712, 'garage_cars'] = 2.0
train.loc[1712, 'garage_area'] = 474
train.loc[1712, 'garage_qual'] = 'TA'
train.loc[1712, 'garage_cond'] = 'TA'
```

##### 7.1.1 Histogram

```python
# Define a function to plot histogram of saleprice vs numerical variables 

def histograms(df, x_vars):
    
    # Set the number of rows/columns and figure size for subplots
    nrows = int(np.ceil(len(x_vars)/3)) 
    ncols = 3
    fig, ax = plt.subplots(nrows = nrows, ncols = ncols, sharey = False, figsize = (15,15))
    
    # Change a multi-dimensional array into a contiguous flattened array.
    ax = ax.ravel()  
    
    # Draw the histogram
    for i, var in enumerate(x_vars): 
        sns.histplot(data = df, x = var, kde = False, ax = ax[i])
        ax[i].set_title(var, fontsize = 13)
        ax[i].set_xlabel('')
        ax[i].tick_params(labelsize=10)
        
    
    # Hide unused plots
    for i in range(nrows * ncols - len(x_vars)):
        ax[-i-1].set_visible(False) # Start from -1
        
    plt.suptitle('Histograms of Numerical Variables', fontsize = 16, y = 1.015)
    
    plt.tight_layout() 
```

```python
# Plot histograms of first 12 numerical variables

histograms(train, num_vars_1)
```

```python
# Define a function to plot scatterplot of saleprice vs numerical variables 

def scatterplot(df, x_vars, target):
    
    # Set the number of rows/columns and figure size for subplots
    nrows = int(np.ceil(len(x_vars)/3)) 
    ncols = 3
    fig, ax = plt.subplots(nrows = nrows, ncols = ncols, sharey = False, figsize = (15,15))
    
    # Change a multi-dimensional array into a contiguous flattened array.
    ax = ax.ravel()  
    
    # Draw the scatterplot
    for i, var in enumerate(x_vars):
        
        # Set the scatterplot properties
        g = sns.regplot(data = df, x = var, y = target, ax = ax[i], marker = 'o', ci = None, color = 'blue', line_kws={'color': 'blue'}, scatter_kws={'s':3})
                
        # Set the scatterplot title, x-label & y_label
        ax[i].set_title(var, fontsize = 13)
        ax[i].set_xlabel('')
        
    # Hide unused plots
    for i in range(nrows * ncols - len(x_vars)):
        ax[-i-1].set_visible(False) # Start from -1
    
    plt.suptitle('Scatterplot of Saleprice vs Numerical Variables', fontsize = 16, y = 1.015)
    
    plt.tight_layout()
```

#### 7.2.1 Boxplots

```python
# Define a function to plot boxplot of categorical variables vs saleprice

def boxplot(df, y_vars, target):
    
    # Set the number of rows/columns and figure size for subplots
    nrows = int(np.ceil(len(y_vars)/3)) 
    ncols = 3
    fig, ax = plt.subplots(nrows = nrows, ncols = ncols, figsize = (20,16))
    
    # Change a multi-dimensional array into a contiguous flattened array.
    ax = ax.ravel()  
    
    # Draw the boxplot
    for i, var in enumerate(y_vars):
    
        # Set the boxplot properties
        g = sns.boxplot(data = df, y = var, x = target, ax = ax[i], orient='h')

        # Set the boxplot title, x-label & y_label
        ax[i].set_title(var, fontsize = 13)
        ax[i].set(ylabel = None)
        
     # Hide unused plots
    for i in range(nrows * ncols - len(y_vars)):
        ax[-i-1].set_visible(False) # Start from -1

    plt.suptitle('Sale Price vs Categorical Variables', fontsize=30, y=1.015)   
    
    plt.tight_layout();
```
