### 5. Model Preparation

#### 5.1 Train Dataset

##### 5.1.1 Linear Regression

```python
# Create our variable matrix (X) and target vector (y) from train dataset

X = train.drop(columns = 'saleprice')
y = train['saleprice']
```

```python
# Perform a split of our X and y 

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 123)
```

```python
# Scale our data to standardize different feature magnitudes.

ss = StandardScaler()
Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)
```

```python
# Instantiate and fit the model.

lr = LinearRegression()
lr.fit(Z_train, y_train)
```

```python
# Train score

lr.score(Z_train, y_train)
```

```python
# Test score

lr.score(Z_test, y_test)
```

##### 5.1.2 Ridge Regression

```python
# Instantiate and fit the model.

ridge = RidgeCV(alphas=np.linspace(1, 200, 100))

ridge.fit(Z_train,y_train)
```

```python
# Optimal ridge alpha

best_ridge_alpha = ridge.alpha_
best_ridge_alpha
```

```python
# Create new ridge model with optimal alpha and fit to training data

ridge = Ridge(alpha=best_ridge_alpha)

ridge.fit(Z_train,y_train)
```

```python
ridge.score(Z_train, y_train)
```

```python
ridge.score(Z_test, y_test)
```

##### 5.1.3 Lasso Regression

```python
# Instantiate and fit the model.

lasso = LassoCV(n_alphas=100)

lasso.fit(Z_train,y_train)
```

```python
# Optimal lasso alpha

best_lasso_alpha = lasso.alpha_
best_lasso_alpha
```

```python
# Create new lasso model with optimal alpha and fit to training data

lasso = Lasso(alpha=best_lasso_alpha)

lasso.fit(Z_train, y_train)
```

```python
lasso.score(Z_train, y_train)
```

```python
lasso.score(Z_test, y_test)
```

```python
# Create scatterplot to show predicted values versus actual values

lasso_preds = lasso.predict(Z_test)

plt.figure(figsize=(10,8))
sns.regplot(data= Z_train, x = np.exp(lasso_preds), y = np.exp(y_test), marker = 'x', color = 'skyblue', line_kws = {'color':'black'})
plt.xlabel('Predicted Sale Price', fontsize=14)
plt.ylabel('Actual Sale Price', fontsize=14)
plt.title('Lasso Predictions of Sale Price vs Actual Sale Price', fontsize=22)
```

```python
ax = sns.scatterplot(data = train, x = y_test, y = y_test-lasso_preds)
ax.axhline(y=0, c='red')
plt.xlabel('Predicted Sale Price', fontsize=12)
plt.ylabel('Actual Sale Price', fontsize=12)
plt.title('Residuals', fontsize=18)
```

### 6. Submission

```python
# Refit model on entire training dataset

X_ss = ss.fit_transform(X)
lasso = LassoCV()
lasso.fit(X_ss, y)
lasso.score(X_ss, y)
```

```python
final_test_scaled = ss.transform(test)
```

```python
final_predictions = lasso.predict(final_test_scaled)
```

```python
test['saleprice'] = np.exp(final_predictions)
```
